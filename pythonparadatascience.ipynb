```python
import matplotlib.pyplot as plt

notas_matematica = ['Matemática',8,7,6,6,7,7,8,10]
notas_portugues = ['Português',9,9,9,8,5,6,8,5]
notas_geografia = ['Geografia',10,10,6,7,7,7,8,7]

notas = [notas_matematica, notas_portugues, notas_geografia]
```

```python
for nota in notas:
 x = list(range(1, 9))
 y = nota[1:]
 plt.plot(x, y, marker='o')
 plt.xlabel('Provas')
 plt.ylabel('Notas')
 plt.title(nota[0])
 plt.show()
```

```python
for nota in notas:
    disciplina = nota[0]
    notas_disciplina = nota[1:]
    plt.barh(disciplina, notas_disciplina)
plt.show()
```

```python
fig, ax = plt.subplots()
plt.figure(figsize=(20,10),dpi=300)
colors = ['red', 'green', 'blue']
for i, nota in enumerate(notas):
    x = list(range(1, 9))
    y = nota[1:]
    ax.plot(x, y, marker='o', label=nota[0], color=colors[i])
    
    # configurando rótulos e título
ax.set_xlabel('Provas')
ax.set_ylabel('Notas')
ax.set_title('Notas por disciplina')


# adicionando legenda
ax.legend()

plt.tight_layout()

# exibindo o gráfico
plt.show()
```
